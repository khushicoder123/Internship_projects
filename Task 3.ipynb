```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].hist(bins=15,edgecolor='black',figsize=(12,6))
```

```python
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=data,x='Species',palette='viridis')
plt.title('Distribution of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SepalLengthCm', y='PetalLengthCm', hue='Species', data=data, palette='viridis')
plt.title('Comparison between Sepal Length and Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Species', y='SepalLengthCm', data=data, palette='viridis')
plt.title('Boxplot of Sepal Length across Species')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
sns.pairplot(data, hue='Species', palette='viridis', markers=["o", "s", "D"])
plt.suptitle('Pairwise Comparisons between Numerical Columns', y=1.02)
plt.show()
```
